구분을 시점명과 종점명으로 나누기

```python
assort = "봉원고가차도->독립문역"
assort
```

```python
def parse_start(assort) :
  start = assort.split("->")[0]
  return str(start)

y = parse_start(assort)
print(type(y))
y
```

```python
def parse_fin(assort) :
  fin = assort.split("->")[-1]
  return str(fin)

parse_fin(assort)
```

```python
traffic_del["시점명"] = traffic_del["구분"].apply(parse_start)
traffic_del.head()
```

```python
traffic_del
```

```python
traffic_del["종점명"] = traffic_del["구분"].apply(parse_fin)
traffic_del.head()
```

```python
pd.merge(left=traffic_del,right=mat_del,how='left',\
         on = ['시점명', '종점명'])
```
